En el juego de cartas póquer, una mano consta de cinco cartas y se clasifican, de menor a mayor, de la siguiente manera:

- Carta Alta: Carta de mayor valor.
- Par: Dos cartas del mismo valor.
- Doble Par: Dos pares diferentes.
- Trío: Tres cartas del mismo valor.
- Escalera: Todas las cartas son valores consecutivos.
- Color: Todas las cartas del mismo palo.
- Full: Trío y un par.
- Póquer: Cuatro cartas del mismo valor.
- Escalera de Color: Todas las cartas son valores consecutivos del mismo palo.
- Escalera Real: Diez, Jota, Reina, Rey, As, del mismo palo.

Las cartas se valoran en el orden:

2, 3, 4, 5, 6, 7, 8, 9, 10, Jota, Reina, Rey, As.
